##The basics

The basic function on which everything else is built is `map_poly_shp`. This takes the path of a shapefile and creates a `PatchCollection` of polygons that may then be tweaked and displayed in Matplotlib using some of the functionality familiar to the Matplotlib user. As a consequence, this is pretty barebone so it's not meant for the final user, but if you know what you're doing, this offers a lot of flexibility.

As a first example, let's externally setup a `figure`, `subplot` and then create the map object, which we add to the figure and display:

```python
fig = figure()
ax = fig.add_subplot(111)
map_obj = maps.map_poly_shp_lonlat(shp_link, projection='merc')
ax.add_collection(map_obj)
ax.set_title('Test')
show()
```

Because often times, all we want to do is to plot the shapefile to make sure nothing is wrong, it'd be useful to have a function that does just that in only one line, instead of having to deal with the figure specs all the time. That's `plot_poly_lines`:

```python
maps.plot_poly_lines(shp_link)
```

The function includes an argument (`savein`) which allows you to not plot the map and instead dump it into a png. You can also modify the color of the polygons (default to transparent) with the argument `poly_col`. Pass the path to the output file and it'll behave this way (commented out not to create a new file, just to illustrate):

```python
#maps.plot_poly_lines(shp_link, savein='my_file.png', poly_col='red')
```

If you just want a quick and dirty way to plot up your shapefile, `plot_poly_lines` is a good way to go; oppositely, if you need more customization, you'll probably have to create the plot yourself and, in that case, `map_poly_shp_lonlat` will be helpful.

##Choropleths

The way in which the architecture for choropleths is built in `PySAL` is very much the same as for basic shapefile plots. It keeps everything as modular as possible to allow for both a great deal of flexibility to the experienced user and ease of use and simplicity for the end-user. To cover both extremes, the module builds on base functions that generate `PatchCollection` objects with barebone choropleths that may be further tinkered if desired, and then offers wrappers that create a "good enough" final figure to quickly and easily display the maps. As before, if you need a lot of customization, the base functions are the way to go, if you are happy with the defaults PySAL offers and want a quick output, use the wrappers.

Currently, the following methods are supported:
    
    - 'classless': maps where every value gets a different color based on a continuous scale
    - 'unique_values': every unique value gets a random value (useful for categorical data)
    - 'quantiles': quantile classification
    - 'fisher_jenks': Fisher-Jenks classification
    - 'equal_interval': equal intervals

Let's first show how you would use the base functions, building simple plots by ourselves, and then we will illustrate the use of the wrapper function `plot_choropleth`. For the former, we'll build the plot in the same way as we did before.

###Classless map

```python
fig = figure()
ax = fig.add_subplot(111)
map_obj = maps.base_choropleth_classless(shp_link, values)
ax.add_collection(map_obj)
ax.set_title('Classless')
show()
```

###Unique values map
For this one, we'll plot an artificial binary variable.

```python
bry = copy(values)
bry[: values.shape[0]/2] = 1
bry[values.shape[0]/2: ] = 0

fig = figure()
ax = fig.add_subplot(111)
map_obj = maps.base_choropleth_unique(shp_link, bry)
ax.add_collection(map_obj)
ax.set_title('Unique values')
show()
```

###Equal interval

```python
fig = figure()
ax = fig.add_subplot(111)
map_obj = maps.base_choropleth_classif(shp_link, values, classification='equal_interval')
ax.add_collection(map_obj)
ax.set_title('Equal interval')
show()
```

###Fisher-Jenks

```python
fig = figure()
ax = fig.add_subplot(111)
map_obj = maps.base_choropleth_classif(shp_link, values, classification='fisher_jenks')
ax.add_collection(map_obj)
ax.set_title('Fisher-Jenks')
show()
```

### Wrapper function
Once we know what everything is based on, let's learn about `plot_choropleth`, which is a wrapper around the previous methods that allows for quick plotting. It offers a basic layout for a map that is simple yet elegant. The idea behind it is pretty similar to that of `plot_poly_lines`, but introducing the choropleth feature. Parallel to it, if you want you can pass a path to a png file and it'll write the figure instead of plotting it. Let's get an example of each of the types supported:

```python
types = ['classless', 'unique_values', 'quantiles', 'fisher_jenks', 'equal_interval']
for typ in types:
    maps.plot_choropleth(shp_link, values, typ, title=typ)
```
